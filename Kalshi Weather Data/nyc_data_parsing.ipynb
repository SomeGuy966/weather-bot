```python
dfs = []

for year in range(2015, 2026):
    df = pd.read_csv(f"nyc_2015-2025/nyc_{year}.csv")

    dfs.append(df)
```

```python
data = pd.concat(dfs, ignore_index=True)
```

```python
data.rename(columns={'TMP' : 'Temperature (C)',
                     'CIG' : 'Cloud Ceiling Height (ft)', 
                     'VIS' : 'Visibility (m)', 
                     'DEW' : 'Dew Point Temperature (C)',
                     'SLP' : 'Sea-level pressure (hPa)',
                     'WND' : 'Wind (m/s)'},
                     inplace=True)
```

```python
data = data[data['Temperature (C)'] != '+9999,9'].reset_index(drop=True)
```

```python
### Removing invalid data points from the rows and setting them to NaN

# Removing invalid data from the wind column
parts = data['Wind (m/s)'].str.split(',', expand=True)
```

```python
# Setting all invalid data points to NaN

parts[parts.iloc[:,0] == '999'] = np.nan
```
